### Analisis Korelasi

```python
# Mencek korelasi antara variabel polutan dengan variabel yang berhubungan dengan cuaca / weather
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
weather_vars = ['TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

print("Correlation Analysis:")
correlation_variables = pollutants + weather_vars
correlation_matrix = tiantan_df[correlation_variables].corr(method="pearson").round(2)
correlation_matrix
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana polusi udara bervariasi setiap musim, dan musim mana yang memiliki kualitas udara terburuk?

```python
# Mendefinisikan musim berdasarkan bulan
tiantan_df['season'] = pd.cut(
    tiantan_df['month'], 
    bins=[0, 3, 6, 9, 12], 
    labels=['Winter', 'Spring', 'Summer', 'Fall'],
    include_lowest=True
)
```

```python
# Menghitung rata-rata harian untuk tiap polutan
daily_avg = tiantan_df.groupby(['year', 'month', 'day', 'season'])[pollutants].mean().reset_index()
```

```python
# Membuat plot untuk pola musiman
plt.figure(figsize=(16, 10))

for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='season', y=pollutant, data=daily_avg)
    plt.title(f'Seasonal Distribution of {pollutant}')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Seasonal Variation of Air Pollutants', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Menghitung rata-rata AQI (Air Quality Index) berdasarkan musim
season_aqi = daily_avg.groupby('season')[pollutants].mean().reset_index()
print("\nAverage Pollutant Levels by Season:")
print(season_aqi.round(2))
```

### Pertanyaan 2: Apa hubungan antara kondisi cuaca dengan kualitas udara?

```python
# Membuat grafik heatmap dari korelasi antara variabel polutan dengan cuaca
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Between Pollutants and Weather Variables')
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Membuat scatter plot untuk hubungan-hubungan yang penting
weather_impact_pairs = [
    ('TEMP', 'PM2.5'),
    ('RAIN', 'PM10'),
    ('WSPM', 'PM2.5'),
    ('WSPM', 'PM10')
]

plt.figure(figsize=(16, 12))
for i, (weather, pollutant) in enumerate(weather_impact_pairs, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=weather, y=pollutant, data=tiantan_df, alpha=0.3)
    
    # Add trend line
    sns.regplot(x=weather, y=pollutant, data=tiantan_df, scatter=False, color='red')
    
    plt.title(f'{weather} vs {pollutant}')
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Weather Variables Impact on Air Pollutants', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Analisis arah angin
plt.figure(figsize=(14, 10))
wind_pollution = tiantan_df.groupby('wd')[pollutants].mean().reset_index()

for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    sns.barplot(x='wd', y=pollutant, data=wind_pollution)
    plt.title(f'Average {pollutant} by Wind Direction')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Impact of Wind Direction on Air Pollutants', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
print("\nCorrelation between Weather Variables and Pollutants:")
for weather in weather_vars:
    correlations = [correlation_matrix.loc[weather, pollutant] for pollutant in pollutants]
    print(f"{weather}: {dict(zip(pollutants, correlations))}")
```

### Pertanyaan 3: Apa pola polusi udara yang terjadi setiap jam dan setiap hari, dan bagaimana informasi ini membantu untuk melakukan perencanaan aktivitas luar ruangan?

```python
# Pola setiap jam
hourly_avg = tiantan_df.groupby('hour')[pollutants].mean()

plt.figure(figsize=(16, 12))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    hourly_avg[pollutant].plot()
    plt.title(f'Hourly Pattern of {pollutant}')
    plt.xlabel('Hour of Day')
    plt.ylabel('Concentration')
    plt.xticks(range(0, 24, 2))
    plt.grid(True)
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Hourly Patterns of Air Pollutants', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Pola setiap hari (hari dalam seminggu)
tiantan_df['day_of_week'] = tiantan_df['datetime'].dt.dayofweek
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
tiantan_df['day_name'] = tiantan_df['day_of_week'].map(dict(zip(range(7), day_names)))

daily_patterns = tiantan_df.groupby('day_name')[pollutants].mean()
daily_patterns = daily_patterns.reindex(day_names)  # Ensure correct order

plt.figure(figsize=(16, 12))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    sns.barplot(x=daily_patterns.index, y=daily_patterns[pollutant])
    plt.title(f'Daily Pattern of {pollutant}')
    plt.xlabel('Day of Week')
    plt.ylabel('Concentration')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Daily Patterns of Air Pollutants', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Mengidentifikasi waktu terbaik untuk aktivitas luar ruangan
print("\nBest Hours for Outdoor Activities (Lowest Pollution):")
best_hours = hourly_avg[['PM2.5', 'PM10', 'O3']].idxmin()
print(best_hours)

print("\nWorst Hours for Outdoor Activities (Highest Pollution):")
worst_hours = hourly_avg[['PM2.5', 'PM10', 'O3']].idxmax()
print(worst_hours)

print("\nBest Days for Outdoor Activities (Lowest Pollution):")
best_days = daily_patterns[['PM2.5', 'PM10', 'O3']].idxmin()
print(best_days)
```

### Pertanyaan 4: Apakah kualitas udara membaik atau memburuk dalam kurun waktu 5 tahun? (2013-2017) ?

```python
# Menghitung rata-rata tahunan
yearly_avg = tiantan_df.groupby('year')[pollutants].mean()
yearly_std = tiantan_df.groupby('year')[pollutants].std()

# Plot yearly trends
plt.figure(figsize=(16, 12))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    yearly_avg[pollutant].plot(marker='o', linestyle='-', linewidth=2)
    
    # Add error bars (standard deviation)
    plt.fill_between(
        yearly_avg.index,
        yearly_avg[pollutant] - yearly_std[pollutant],
        yearly_avg[pollutant] + yearly_std[pollutant],
        alpha=0.3
    )
    
    plt.title(f'Yearly Trend of {pollutant}')
    plt.xlabel('Year')
    plt.ylabel('Concentration')
    plt.grid(True)
    plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle('Air Quality Trends (2013-2017)', fontsize=16)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Menghitung persentase perubahan dari tahun 2013 ke tahun 2017
pct_change = ((yearly_avg.loc[2017] - yearly_avg.loc[2013]) / yearly_avg.loc[2013] * 100).round(2)
print("\nPercentage Change in Pollutant Levels (2013-2017):")
print(pct_change)
```

```python
# Menghitung jumlah hari yang melewati petunjuk WHO berdasarkan tahun
# WHO guidelines: PM2.5: 25 μg/m³, PM10: 50 μg/m³, NO2: 40 μg/m³, SO2: 20 μg/m³, O3: 100 μg/m³
who_limits = {
    'PM2.5': 25,
    'PM10': 50,
    'NO2': 40,
    'SO2': 20,
    'O3': 100
}

# Menghitung rata-rata harian
daily_data = tiantan_df.groupby(['year', 'month', 'day'])[pollutants].mean().reset_index()

# Menghitung hari yang melewati
exceeding_days = {}
for pollutant, limit in who_limits.items():
    yearly_counts = daily_data[daily_data[pollutant] > limit].groupby('year').size()
    yearly_total = daily_data.groupby('year').size()
    yearly_percentage = (yearly_counts / yearly_total * 100).round(1)
    exceeding_days[pollutant] = yearly_percentage

exceeding_df = pd.DataFrame(exceeding_days)
print("\nPercentage of Days Exceeding WHO Guidelines by Year:")
print(exceeding_df)
```
